1. Determine the distribution of employees across each team and calculate 
the percentage split relative to the total number of employees.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
team_counts=df['Team'].value_counts()
team_counts
```

```python
explode = [0.1 if count == team_counts.max() else 0 for count in team_counts]
```

```python
plt.figure(figsize=(10, 10))
plt.pie(team_counts, autopct='%1.1f%%', labels=team_counts.index, explode=explode)
plt.title('Distribution of Employees Across Teams')
plt.show()
```

2.Second highest group with most employee is Memphis Grizzlies(18) with 3.9% of total employees.

3.77% of teams have same number of employees(15).

4.Discover which team and position have the highest salary expenditure

```python
high_sal_team=df.groupby('Team')['Salary'].sum().reset_index()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(x='Team',height='Salary',data=high_sal_team,color='g')
for i in range(len(high_sal_team)):
    plt.text(high_sal_team['Team'][i], high_sal_team['Salary'][i] + 1, str(high_sal_team['Salary'][i]),
             ha='center',rotation='vertical',color='k')
plt.xticks(rotation=90)
plt.title('Salary Expenditure by Team')
plt.xlabel('Teams')
plt.ylabel('Total Salary Expenditure')
plt.show()
```

1. From this graph we can clearly understand  the team 'Cleveland Cavaliers'  has the highest salary expenditure with 106988689

2. The second highest team with salary expenditure team 'Los Angeles Clippers' with 94854640

3.Thel lowest salary expenditure is team 'Philadelphia 76ers' with 30992894

```python
high_sal_pos=df.groupby('Position')['Salary'].sum().reset_index()
```

```python

plt.bar(x='Position',height='Salary',data=high_sal_pos,color='b')
for i in range(len(high_sal_pos)):
    plt.text(high_sal_pos['Position'][i], high_sal_pos['Salary'][i] + 1, str(high_sal_pos['Salary'][i]),
             ha='center',color='k')
plt.xticks(rotation=90)
plt.title('Salary Expenditure by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary Expenditure')

plt.show()
```
